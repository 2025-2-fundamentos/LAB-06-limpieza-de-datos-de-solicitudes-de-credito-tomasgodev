```python
import pandas as pd
df = pd.read_csv("../files/input/solicitudes_de_credito.csv", sep=";")
print(df.head())
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].astype(str).apply(lambda x: x.strip().lower())


    else:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace("$", "", regex=False)
            .str.replace(",", "", regex=False)
            .str.replace(" ", "", regex=False)
            .str.replace("-", "", regex=False)
        )


print(df.shape[0])
print(df.sexo.value_counts().to_list())


df.drop_duplicates(inplace=True)
print(df.shape[0])
```
